## 3.  NumPysta DataFrameen
Tee NumPy-funktioilla NumPy-taulukko jossa on kokonaisluvut 1:sta 99:een.  
Tee edellisen taulukon perusteella NumPy-taulukko, jossa on kyseisten lukujen jakojäännökset seitsemällä jaettaessa. 

Tee näistä kahdesta NumPy-taulukosta DataFrame, jossa nämä ovat sarakkeina, 

Tee sitten NumPy:n randint-funktiolla taulukko, jossa on 99 kpl satunnaislukuja väliltä 1-6, ja lisää tämä uutena sarakkeena DataFrameen.
https://docs.scipy.org/doc/numpy-1.15.1/reference/generated/numpy.random.randint.html#numpy.random.randint

Suodata sitten DataFramesta:
* 10 viimeistä riviä
* rivit, joissa jakojäännös on 0 ja satunnaisluku 6
* 3 ensimmäistä riviä niistä, joissa luku on parillinen, jakojäännös < 3 ja satunnaisluku > 4


```python
luvut = np.arange(1,100)
jakoj = luvut % 7

df = pd.DataFrame({'luku':luvut, 'jakoj':jakoj })

satn = np.random.randint(1,7,99)



df["sat"] = satn

print(df.iloc[-10:,:])

print('\n---------\n')

print(df[(df['jakoj'] == 0) & (df['sat']==6)]) 

print('\n---------\n')

print(df[(df['luku'] % 2 == 0) & (df['jakoj']<3) & (df['sat']>4)].iloc[:4,:] )
```
